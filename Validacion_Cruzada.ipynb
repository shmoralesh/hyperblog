### Importamos las librerías necesarias para el análisis

```python
# Vamos a distinguir las variables numéricas de las categóricas
X = df.iloc[:,2:] # Numéricas
Y = df.iloc[:,1] # Categóricas --> Si el cáncer es benigno o maligno
# Ahora la idea es poder explicar la variable Y (si el cáncer es B o M) con las demás variables X
```

### Ahora podemos crear la división de modelos:

```python
# Necesitamos una librería nueva, para árboles de decisión. Así como el modelo de decisión
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 4)
# El random state determina el nivel de precisión que estamos solicitando

modelo = DecisionTreeClassifier()

# Hacemos el fit del modelo, la coincidencia dentro del modelo. Que hacemos entre lo que estamos entrenando
modelo.fit(X_train, Y_train)

# Luego solicitaremos el resultado con score, que es la puntuación el modelo de text de X y Y
resultado = modelo.score(X_test, Y_test)
print(resultado)
# El porcentaje que aparece en resultado nosdice cual es la precisión del modelo en porcentaje. Un 88% de precisión es muy bueno
```

### Aplicación de la validación cruzada k fold:

```python
# Importamos el módulo kfold de la librería
from sklearn.model_selection import KFold
```

```python
modelo = DecisionTreeClassifier()
# Ahora le damos el número de Ks que queremos, lo mas usual es 10
kfold_validacion = KFold(10)
```

```python
# Conclusión: Buscamos que de las 10 folds(muestras) vamos a sacar los promedios. Esto es para hacer un cruce de validaciones donde no partimos de una, sino de 10 observaciones.
# SObre estas 10 validaciones sacamos el promedio para no sesgarnos y para ver que el modelo funciona

# Importamos el modulo para cross validation score
from  sklearn.model_selection import cross_val_score
```

```python
resultados = cross_val_score(modelo, X, Y, cv = kfold_validacion)
print(resultados)
resultados.mean()

# Así que vemos el resultado para cada uno de los 10 folds que solicitamos. Viendo que es un modelo muy bueno
```
