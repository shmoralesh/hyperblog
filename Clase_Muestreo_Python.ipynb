#### Muestreo sistemático: Tiene una norma, o intervalo regular

```python
# Se define una función usando econdata
def systematic_sampling(econdata,step):
  # Se definen los índices que van desde el cero hasta la longitud de econdata, para integrar los posibles valores
  # El step, o la función de intervalos la definiremos a continuación
  # Los indexes modifican: donde empieza la selección, donde termina, y de cuanto es cada salto
  indexes = np.arange(0,len(econdata),step=step)
  systematic_sample = econdata.iloc[indexes]
  # y se devuelve el valor de systematic sample
  return systematic_sample

# Llamamos la función y le decimos que seleccione datos de 3 en 3
systematic_sample = systematic_sampling(econdata,3)
systematic_sample
```

#### 3) Muestreo estratificado

```python
# Creamos la variable de estratificación

# Creamos una columna juntando el lugar y el sitio y después contamos cuantos datos están en cada uno. Lo dividimos entrela cantidad de datos de la BD (len(econdata))
# Se organizan en orden descendente (sort_values(ascending=False))
econdata["estratificado"] = econdata["delegacion"] + "," + econdata["tipo"]
(econdata["estratificado"].value_counts()/len(econdata)).sort_values(ascending=False)
```

Nuestros datos dicen que la proporción es la siguiente:

1) hoteles en Cuauhtemoc 0.5

2) Museos en Cuauhtémoc 0.2

3) Hoteles en VEnustiano Carranza 0.1

4) Mercados en Cuauhtémoc 0.1

5) Mercados en Venistiano Carranza 0.1

```python
# Crear una nueva función con: econdata, los nombres de las columnas estratificadas, los valores estratificados, las proporciones estratificadas, y el random state = None
def data_estratificada(econdata,nombres_columnas_estrat,valores_estrat,prop_estrat,random_state=None):
  # Vamos a crear un data frame vacío con los nombre de columnas son las columnas de econdata
  df_estrat = pd.DataFrame(columns = econdata.columns)

  # Nuestra posición inicia en -1
  pos = -1
  # Creamos un ciclo for que va desde i hasta la longitud de los valores estratificados
  for i in range(len(valores_estrat)):
    # Le vamos sumando 1 a cada posición
    pos += 1
    # Y si la posición llega a ser igual a la logitud de los valores estratificados menos 1 valor
    if pos == len(valores_estrat) - 1:
      # a continuación definiremos la ratio de la longitud, que será la resta de la longitud de econdata menos la longitud de la nueva dataframe
      # la idea es que se esté comparando todo el tiempo
      ratio_len = len(econdata)- len(df_estrat)
    else:
      # De lo contrario definimos un nuevo ratio, y le diremos que es un entero de la longitud de econdata y lo equivale  la proporción estratificada que vamos a definir, para cada uno de los i
      # Esto nos ayuda a calcular cual es cada una de las filas, el componente i, y a sacar el ratio en basa a cada una de la proporción seleccionada
      ratio_len = int(len(econdata) * prop_estrat[i])

    # Luego creamos el data frame filtrado, a partir de la tabla de econdata, extrayendo los nombres de las columnas estratificadas y dándole los valores para cada elemento i
    df_filtrado = econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]]
    df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)

    # Se modifica el data frame vacio pegandole los valores
    df_estrat = pd.concat([df_estrat, df_temp])

  # Se devuelve el valor de la variable modificada
  return df_estrat
```

```python
# Se inicializan los valores necesarios para correr la función
valores_estrat = ["Cuautémoc,Hotel","Cuautémoc,Museo","Venustiano Carranza,Hotel","Cuauhtémoc,Mercado","Venustiano Carranza,Mercado"]
# Se colocan las proporciones que habíamos definido
prop_estrat = [0.5,0.2,0.1,0.1,0.1]

# Iniciamos la función con los datos que habíamos definido
df_estrat = data_estratificada(econdata,"estratificado",valores_estrat,prop_estrat,random_state=42)
df_estrat
```
